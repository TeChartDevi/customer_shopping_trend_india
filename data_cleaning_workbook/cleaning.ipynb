Feature Engineering:
- 1. Creating a New Region Column.
- 2. Extracting Parts of date.
- 3. Creating Age group 

- 1. Creating a New Region Column:

```python
# All unique city names:

df['Cities'].unique()
```

```python
# Mapping the cities

region_map = {
    'Delhi': 'North',
    'Noida': 'North',
    'Gurgaon': 'North',
    'Chandigarh': 'North',
    'Jaipur': 'North',
    'Lucknow': 'North',

    'Chennai': 'South',
    'Hyderabad': 'South',
    'Visakhapatnam': 'South',
    'Kochi': 'South',
    'Bangalore': 'South',

    'Mumbai': 'West',
    'Pune': 'West',
    'Ahmedabad': 'West',
    'Nagpur': 'West',
    'Surat': 'West',

    'Kolkata': 'East',
    'Patna': 'East',
    'Bhubaneswar': 'East',

    'Indore': 'Central'
}

df['Region'] = df['Cities'].map(region_map)
```

```python
# Popping out the region column:

region = df.pop('Region')

# Finding the position of the Cities column:

loc = df.columns.get_loc('Cities')

# Inserting Region column immediately after Cities column:

df.insert(loc + 1, 'Region', region)
```

- 2. Extracting Parts of date and inserting them immediately after the Purchase Date column:

```python
# Finding the position of the Purchase Date column:
loc = df.columns.get_loc('Purchase Date')


# Extracting Year and inserting it immediately after the Purchase Date column:
df.insert(loc + 1, 'Year', df['Purchase Date'].dt.year)


# Extracting Month and inserting it immediately after the Year column:
df.insert(loc + 2, 'Month', df['Purchase Date'].dt.month)


# Extracting Day and inserting it immediately after the Month column:
df.insert(loc + 3, 'Day', df['Purchase Date'].dt.day)


# Extracting Weekday Name and inserting it immediately after the Day column:
df.insert(loc + 4, 'Weekday', df['Purchase Date'].dt.day_name())
```

- 3. Creating Age group:

```python
def age_group(age):
    if age < 18:
        return 'Under 18'
    elif age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 60:
        return '46-60'
    else:
        return 'Above 60'
    
# Finding the position of the age column:
loc = df.columns.get_loc('Age')

# Inserting the Age Group column immediately after the age column:
df.insert(loc + 1, 'Age Group', df['Age'].apply(age_group))
```
